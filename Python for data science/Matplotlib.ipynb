# Pyplot:

        --> Most of the Matplotlib utilites lies under the pyplot submodule and are usually imported under the plt alias.

```python
import matplotlib.pyplot as plt
x=([1,3])
y=([3,10])
plt.plot(x,y)
plt.show()
```

# Plotting x any y points:

        --> The plot() function is used to draw points(markers) in a diagram.
        
        --> By default,the plot() function draws a line from point to point.
        
        --> The function takes parameters for specifying points in the diagram.
        
        --> parameter 1 is an array containing the points on the x-axis.
        
        --> parameter 2 is an array containing the points on the y-axis.
        

```python
import matplotlib.pyplot as plt
import numpy as np
x =np.array([2,6,7,9,12])
y = np.array([12,9,7,4,3])
plt.plot(x,y)
plt.show()
```

# Markers:
    
    --> you can use the keyword argument markerr to emphasize each point with a specified marker.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([3,5,8,9,10])
y = np.array([1,8,5,6,4])
plt.plot(x,y,marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([3,6,8,11,14])
y=np.array([0,7,6,18,4])
plt.plot(x,y,marker='*')
plt.show()
```

# Matplotlib line:

# Multiple lines:
    
            --> you can plot as many lines as you like by simply adding mort plt.plot() functions.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,7,3,12])
y= np.array([4,11,0,11])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
x2 = np.array([0,1,2,3])
y2 = np.array([6,2,7,11])

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()
```

# Matplotlib Labels and Titels:

        --> You can use the xLabel() and yLabel() function to set a albel for the x and y axis.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,color='r')

plt.xlabel('points')
plt.ylabel('score')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([18,19,22,23,25,26])
y = np.array([10000,12000,13000,15000,17000,19000])

plt.plot(x,y,color='g')

plt.title('Salary Details')
plt.xlabel('Age')
plt.ylabel('salary')

plt.show()
```

# Add grid lines to a plot:

        --> you can use the grid() function to add grid lines to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,4,5,6,7,9])
y = np.array([3,4,5,6,7,8])

plt.plot(x,y,color='r')

plt.title('score table')
plt.xlabel('points')
plt.ylabel('score')

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([2,4,6,7,8,12])
y= np.array([4,8,12,15,18,21])

plt.title('score board')
plt.xlabel('points')
plt.ylabel('score')

plt.grid(axis = 'x')

plt.plot(x,y,color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([2,4,6,7,8,12])
y= np.array([4,8,12,15,18,21])

plt.title('score board')
plt.xlabel('points')
plt.ylabel('score')

plt.grid(axis = 'y')

plt.plot(x,y,color='g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([2,4,6,7,8,12])
y= np.array([4,8,12,15,18,21])

plt.title('score board')
plt.xlabel('points')
plt.ylabel('score')

plt.grid(color='g',linestyle='--')

plt.plot(x,y,color='r')
plt.show()
```

# Creating bars using matplotlib:

        --> you can usee the bar functuin to draw bar graphs.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y,color='r')
plt.show()
```

# Histogram in matplotlib:

        --> Histogram is a graph =showing frequency distribution.
        
        --> the graph showing the number of observations within each given interval.

```python
import numpy as np

x= np.random.normal(170,10,250)

print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.random.normal(17,10,250)

plt.hist(x,color='r')
plt.show()
```
